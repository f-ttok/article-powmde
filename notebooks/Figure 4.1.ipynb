```julia
using PyPlot

rcParams = PyPlot.PyDict(PyPlot.matplotlib."rcParams")
rcParams["font.size"] = 10
```

```julia
function d0(λ)
    μ = log(λ)^2 + 5*π^2/4
    c = sqrt(μ - sqrt(μ^2 - π^4)) / sqrt(π^2/2)
    return asin(c)
end
```

```julia
λ = 10.0 .^ LinRange(-16, 16, 201)
d = d0.(λ)

fig, ax = subplots(1, 1, figsize=[7, 5]/2.54, dpi=150)

ax.plot([1e-16,1e16], [π/2, π/2], c="k", lw=1, ls="--")
ax.plot(λ, d, c="k", lw=2)
ax.set_xscale("log")
ax.set_xticks([1e-16, 1e-8, 1e0, 1e8, 1e16])
labels = ["\$10^{$(k)}\$" for k = -16:8:16]
labels[3] = "1"
ax.set_xticklabels(labels)
ax.set_xlim(1e-16,1e16)
ax.set_xlabel("\$\\lambda\$")
ax.set_yticks([0, π/4, π/2])
ax.set_yticklabels(["0", "\$\\pi/4\$", "\$\\pi/2\$"])
ax.set_ylabel("\$d_0(\\lambda)\$")

fig.tight_layout()
fig.savefig("fig-d0.pdf")

show()
```
